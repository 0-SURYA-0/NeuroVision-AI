```python
import sys
import os
import time
import warnings
from pathlib import Path
from typing import Dict, List, Tuple, Optional, Union, Any
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from collections import defaultdict
import pickle

# Add 3D architecture to path
sys.path.append(r"c:\Users\surya\NueroVisionAI\core\3D architecture")

# Import 3D modules
from utils_3d import Config, setup_logging, set_random_seed, get_device, monitor_memory, timer, format_time
from model_3d import UNet3D, create_unet3d
from dataset_3d import BraTSDataset, create_data_loaders, DataAugmentation3D
from train_3d import ModelTrainer, create_trainer, DiceLoss, FocalLoss, ComboLoss, calculate_dice_score

print("🚀" * 50)
print("ADVANCED 3D BRAIN TUMOR SEGMENTATION TRAINING")
print("RTX 4060 GPU OPTIMIZED PIPELINE")
print("🚀" * 50)

# GPU Detection and Setup
device = get_device("auto")
print(f"\n🔥 GPU STATUS:")
print(f"├── Device: {device}")
if torch.cuda.is_available():
    print(f"├── GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"├── CUDA Version: {torch.version.cuda}")
    print(f"├── PyTorch Version: {torch.__version__}")
    print(f"├── Memory Total: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
    print(f"└── Mixed Precision: ENABLED ✅")
else:
    print(f"└── CPU Mode: Enabled")
    
# Set random seed for reproducibility
set_random_seed(42, deterministic=True)

# Configuration
config = Config()
config.update(
    # Data settings
    data_dir=r"c:\Users\surya\NueroVisionAI\data\BrainWithTumor",
    input_size=(64, 64, 64),  # Memory-efficient size for RTX 4060
    
    # Training settings optimized for RTX 4060
    batch_size=1,
    num_workers=0,  # Windows-safe
    pin_memory=True,
    mixed_precision=True,
    
    # Model settings
    model_config="light",
    base_filters=16,
    depth=3,
    dropout_rate=0.1,
    
    # Training parameters
    learning_rate=1e-4,
    num_epochs=20,
    patience=10,
    
    # Loss function
    loss_function="combo",
    dice_weight=0.7,
    focal_weight=0.3,
    
    # Paths
    checkpoint_dir=r"c:\Users\surya\NueroVisionAI\checkpoints\3d",
    output_dir=r"c:\Users\surya\NueroVisionAI\outputs\3d", 
    log_dir=r"c:\Users\surya\NueroVisionAI\logs\3d",
    
    # Hardware
    device="auto"
)

print(f"\n🔧 TRAINING CONFIGURATION:")
print(f"├── Model: {config.model_config} 3D U-Net")
print(f"├── Volume Size: {config.input_size}")
print(f"├── Batch Size: {config.batch_size} (RTX 4060 optimized)")
print(f"├── Epochs: {config.num_epochs}")
print(f"├── Learning Rate: {config.learning_rate}")
print(f"├── Mixed Precision: {'✅ ENABLED' if config.mixed_precision else '❌ DISABLED'}")
print(f"└── Loss Function: {config.loss_function.upper()}")

# Create directories
for dir_path in [config.checkpoint_dir, config.output_dir, config.log_dir]:
    Path(dir_path).mkdir(parents=True, exist_ok=True)

print(f"\n📁 DIRECTORIES CREATED:")
print(f"├── Checkpoints: {config.checkpoint_dir}")
print(f"├── Outputs: {config.output_dir}")
print(f"└── Logs: {config.log_dir}")

# Memory Monitoring Class
class GPUMemoryMonitor:
    def __init__(self, device):
        self.device = device
        
    def get_memory_info(self):
        if self.device.type == 'cuda':
            allocated = torch.cuda.memory_allocated(self.device) / 1024**2
            cached = torch.cuda.memory_reserved(self.device) / 1024**2
            return f"{allocated:.0f}MB allocated, {cached:.0f}MB cached"
        return "CPU mode"
    
    def clear_cache(self):
        if self.device.type == 'cuda':
            torch.cuda.empty_cache()

memory_monitor = GPUMemoryMonitor(device)

# Advanced Training Visualizer
class AdvancedTrainingVisualizer:
    def __init__(self):
        self.train_losses = []
        self.val_losses = []
        self.train_dice = []
        self.val_dice = []
        self.learning_rates = []
        self.epochs = []
        
    def update(self, epoch, train_loss, val_loss, train_dice_score, val_dice_score, lr):
        self.epochs.append(epoch)
        self.train_losses.append(train_loss)
        self.val_losses.append(val_loss)
        self.train_dice.append(train_dice_score)
        self.val_dice.append(val_dice_score)
        self.learning_rates.append(lr)
        
    def plot_training_progress(self, save_path=None):
        if len(self.epochs) < 2:
            return
            
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('🧠 3D Brain Tumor Segmentation Training Progress', fontsize=16, fontweight='bold')
        
        # Loss curves
        ax1.plot(self.epochs, self.train_losses, 'b-', label='Training Loss', linewidth=2)
        ax1.plot(self.epochs, self.val_losses, 'r-', label='Validation Loss', linewidth=2)
        ax1.set_title('📉 Loss Curves', fontweight='bold')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # Dice score curves
        ax2.plot(self.epochs, self.train_dice, 'b-', label='Training Dice', linewidth=2)
        ax2.plot(self.epochs, self.val_dice, 'r-', label='Validation Dice', linewidth=2)
        ax2.set_title('🎯 Dice Score Curves', fontweight='bold')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Dice Score')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Learning rate
        ax3.plot(self.epochs, self.learning_rates, 'g-', linewidth=2)
        ax3.set_title('📈 Learning Rate Schedule', fontweight='bold')
        ax3.set_xlabel('Epoch')
        ax3.set_ylabel('Learning Rate')
        ax3.set_yscale('log')
        ax3.grid(True, alpha=0.3)
        
        # Performance summary
        if self.val_dice:
            best_dice = max(self.val_dice)
            best_epoch = self.epochs[self.val_dice.index(best_dice)]
            
            ax4.text(0.1, 0.8, f'🏆 BEST PERFORMANCE', fontsize=14, fontweight='bold', transform=ax4.transAxes)
            ax4.text(0.1, 0.7, f'Best Dice Score: {best_dice:.4f}', fontsize=12, transform=ax4.transAxes)
            ax4.text(0.1, 0.6, f'Best Epoch: {best_epoch}', fontsize=12, transform=ax4.transAxes)
            ax4.text(0.1, 0.5, f'Current Epoch: {self.epochs[-1]}', fontsize=12, transform=ax4.transAxes)
            ax4.text(0.1, 0.4, f'Current Val Dice: {self.val_dice[-1]:.4f}', fontsize=12, transform=ax4.transAxes)
            ax4.text(0.1, 0.3, f'🔥 RTX 4060 Optimized', fontsize=12, fontweight='bold', color='red', transform=ax4.transAxes)
            
        ax4.set_xlim(0, 1)
        ax4.set_ylim(0, 1)
        ax4.axis('off')
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.show()

visualizer = AdvancedTrainingVisualizer()

# Custom Dataset for Quick Testing
class QuickBrainDataset(Dataset):
    def __init__(self, size=100, input_size=(64, 64, 64)):
        self.size = size
        self.input_size = input_size
        
    def __len__(self):
        return self.size
        
    def __getitem__(self, idx):
        # Generate synthetic brain-like data with tumor
        volume = torch.randn(4, *self.input_size) * 0.1  # 4 modalities
        
        # Add brain-like structure
        center = [s // 2 for s in self.input_size]
        for i in range(4):
            for z in range(self.input_size[0]):
                for y in range(self.input_size[1]):
                    for x in range(self.input_size[2]):
                        dist = ((z - center[0])**2 + (y - center[1])**2 + (x - center[2])**2) ** 0.5
                        if dist < min(self.input_size) * 0.3:  # Brain region
                            volume[i, z, y, x] += 0.5
        
        # Add tumor
        tumor_center = [int(c + np.random.normal(0, 5)) for c in center]
        tumor_size = np.random.randint(3, 8)
        mask = torch.zeros(1, *self.input_size)
        
        for z in range(max(0, tumor_center[0] - tumor_size), min(self.input_size[0], tumor_center[0] + tumor_size)):
            for y in range(max(0, tumor_center[1] - tumor_size), min(self.input_size[1], tumor_center[1] + tumor_size)):
                for x in range(max(0, tumor_center[2] - tumor_size), min(self.input_size[2], tumor_center[2] + tumor_size)):
                    dist = ((z - tumor_center[0])**2 + (y - tumor_center[1])**2 + (x - tumor_center[2])**2) ** 0.5
                    if dist < tumor_size:
                        mask[0, z, y, x] = 1.0
                        # Enhance tumor in volume
                        for i in range(4):
                            volume[i, z, y, x] += 0.3
        
        return volume.float(), mask.float(), {'case_id': f'synthetic_{idx}'}

print(f"\n📊 CREATING DATASETS...")

try:
    # Try to create real data loaders
    data_loaders = create_data_loaders(
        data_dir=config.data_dir,
        config=config,
        splits=["train", "val"]
    )
    train_loader = data_loaders["train"]
    val_loader = data_loaders["val"]
    print(f"✅ Real dataset loaded: {len(train_loader.dataset)} training samples")
    
except Exception as e:
    print(f"⚠️  Real dataset not available: {str(e)}")
    print(f"🔄 Using synthetic dataset for demonstration...")
    
    # Create synthetic datasets
    train_dataset = QuickBrainDataset(size=80, input_size=config.input_size)
    val_dataset = QuickBrainDataset(size=20, input_size=config.input_size)
    
    train_loader = DataLoader(
        train_dataset,
        batch_size=config.batch_size,
        shuffle=True,
        num_workers=config.num_workers,
        pin_memory=config.pin_memory
    )
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=config.batch_size,
        shuffle=False,
        num_workers=config.num_workers,
        pin_memory=config.pin_memory
    )
    
    print(f"✅ Synthetic dataset created: {len(train_dataset)} training samples")

print(f"├── Training batches: {len(train_loader)}")
print(f"├── Validation batches: {len(val_loader)}")
print(f"└── Batch size: {config.batch_size}")

# Create model
print(f"\n🧠 CREATING 3D U-NET MODEL...")
model = create_unet3d(config.model_config).to(device)

# Model info
num_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
model_size_mb = num_params * 4 / (1024 * 1024)  # Assuming float32

print(f"✅ Model created successfully!")
print(f"├── Architecture: {config.model_config.upper()} 3D U-Net")
print(f"├── Total parameters: {num_params:,}")
print(f"├── Trainable parameters: {trainable_params:,}")
print(f"├── Model size: {model_size_mb:.1f} MB")
print(f"├── Input channels: 4 (T1, T1CE, T2, FLAIR)")
print(f"├── Output channels: 1 (tumor segmentation)")
print(f"└── Device: {next(model.parameters()).device}")

# Loss function and optimizer
criterion = ComboLoss(
    dice_weight=config.dice_weight,
    focal_weight=config.focal_weight
)

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=config.learning_rate,
    weight_decay=1e-5
)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='max',
    factor=0.5,
    patience=config.patience // 2,
    verbose=True,
    min_lr=config.learning_rate * 0.01
)

# Mixed precision scaler
scaler = GradScaler() if config.mixed_precision else None

print(f"\n⚙️  TRAINING COMPONENTS:")
print(f"├── Loss function: Combo Loss (Dice + Focal)")
print(f"├── Optimizer: AdamW")
print(f"├── Scheduler: ReduceLROnPlateau")
print(f"├── Mixed precision: {'✅ ENABLED' if scaler else '❌ DISABLED'}")
print(f"└── Memory monitor: Active")

# Training state
best_val_dice = 0.0
best_val_loss = float('inf')
training_history = []
early_stop_counter = 0

print(f"\n🚀 STARTING TRAINING ON RTX 4060!")
print("=" * 80)
print(f"🔥 GPU MEMORY: {memory_monitor.get_memory_info()}")
print("=" * 80)

training_start_time = time.time()

try:
    for epoch in range(config.num_epochs):
        epoch_start_time = time.time()
        
        # Clear GPU cache before each epoch
        memory_monitor.clear_cache()
        
        print(f"\n🔥 EPOCH {epoch + 1}/{config.num_epochs}")
        print(f"{'─' * 60}")
        print(f"📊 GPU Memory: {memory_monitor.get_memory_info()}")
        print(f"📈 Learning Rate: {optimizer.param_groups[0]['lr']:.2e}")
        
        # ============ TRAINING PHASE ============
        model.train()
        train_loss = 0.0
        train_dice = 0.0
        
        print(f"\n🏋️  TRAINING PHASE")
        train_pbar = tqdm(
            train_loader, 
            desc=f"Epoch {epoch+1:02d} Train",
            ncols=100,
            bar_format="{desc}: {percentage:3.0f}%|{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}, {rate_fmt}] {postfix}"
        )
        
        for batch_idx, (inputs, targets, metadata) in enumerate(train_pbar):
            inputs = inputs.to(device, non_blocking=True)
            targets = targets.to(device, non_blocking=True)
            
            optimizer.zero_grad(set_to_none=True)
            
            if scaler:
                with autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, targets)
                
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                outputs = model(inputs)
                loss = criterion(outputs, targets)
                loss.backward()
                optimizer.step()
            
            # Calculate metrics
            with torch.no_grad():
                dice_score = calculate_dice_score(outputs, targets).mean()
            
            train_loss += loss.item()
            train_dice += dice_score.item()
            
            # Update progress bar
            train_pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Dice': f'{dice_score.item():.4f}',
                'GPU': f'{torch.cuda.memory_allocated(device) / 1024**2:.0f}MB' if device.type == 'cuda' else 'CPU'
            })
            
            # Cleanup
            del inputs, targets, outputs, loss
        
        avg_train_loss = train_loss / len(train_loader)
        avg_train_dice = train_dice / len(train_loader)
        
        # ============ VALIDATION PHASE ============
        model.eval()
        val_loss = 0.0
        val_dice = 0.0
        
        print(f"\n🎯 VALIDATION PHASE")
        val_pbar = tqdm(
            val_loader,
            desc=f"Epoch {epoch+1:02d} Val  ",
            ncols=100,
            bar_format="{desc}: {percentage:3.0f}%|{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}] {postfix}"
        )
        
        with torch.no_grad():
            for inputs, targets, metadata in val_pbar:
                inputs = inputs.to(device, non_blocking=True)
                targets = targets.to(device, non_blocking=True)
                
                if scaler:
                    with autocast():
                        outputs = model(inputs)
                        loss = criterion(outputs, targets)
                else:
                    outputs = model(inputs)
                    loss = criterion(outputs, targets)
                
                dice_score = calculate_dice_score(outputs, targets).mean()
                
                val_loss += loss.item()
                val_dice += dice_score.item()
                
                val_pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Dice': f'{dice_score.item():.4f}'
                })
                
                del inputs, targets, outputs, loss
        
        avg_val_loss = val_loss / len(val_loader)
        avg_val_dice = val_dice / len(val_loader)
        
        # Update learning rate scheduler
        scheduler.step(avg_val_dice)
        current_lr = optimizer.param_groups[0]['lr']
        
        # Update visualizer
        visualizer.update(
            epoch + 1, avg_train_loss, avg_val_loss, 
            avg_train_dice, avg_val_dice, current_lr
        )
        
        # Calculate epoch time
        epoch_time = time.time() - epoch_start_time
        
        # Check for improvement
        is_best = avg_val_dice > best_val_dice
        if is_best:
            best_val_dice = avg_val_dice
            best_val_loss = avg_val_loss
            early_stop_counter = 0
            improvement_indicator = "🎉 NEW BEST!"
        else:
            early_stop_counter += 1
            improvement_indicator = ""
        
        # Store training history
        training_history.append({
            'epoch': epoch + 1,
            'train_loss': avg_train_loss,
            'val_loss': avg_val_loss,
            'train_dice': avg_train_dice,
            'val_dice': avg_val_dice,
            'learning_rate': current_lr,
            'epoch_time': epoch_time,
            'is_best': is_best
        })
        
        # Print epoch summary
        print(f"\n{'🎉 ' if is_best else ''}EPOCH {epoch + 1} SUMMARY {improvement_indicator}")
        print(f"{'═' * 60}")
        print(f"📉 Train Loss: {avg_train_loss:.4f} → Val Loss: {avg_val_loss:.4f}")
        print(f"🎯 Train Dice: {avg_train_dice:.4f} → Val Dice: {avg_val_dice:.4f}")
        print(f"⏱️  Time: {epoch_time:.1f}s | LR: {current_lr:.2e}")
        print(f"🔥 GPU: {memory_monitor.get_memory_info()}")
        print(f"🏆 Best Val Dice: {best_val_dice:.4f}")
        
        # Save checkpoint
        if is_best:
            checkpoint = {
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'scheduler_state_dict': scheduler.state_dict(),
                'best_val_dice': best_val_dice,
                'config': config,
                'training_history': training_history
            }
            
            checkpoint_path = Path(config.checkpoint_dir) / "best_model.pth"
            torch.save(checkpoint, checkpoint_path)
            
            # Also save as .pkl for easy loading
            model_path = Path(config.output_dir) / "unet3d_model.pkl"
            torch.save(model.state_dict(), model_path)
            
            print(f"💾 Best model saved: {checkpoint_path}")
            print(f"💾 Model weights saved: {model_path}")
        
        # Plot progress every few epochs
        if (epoch + 1) % 5 == 0 or is_best:
            try:
                plot_path = Path(config.output_dir) / f"training_progress_epoch_{epoch+1:03d}.png"
                visualizer.plot_training_progress(save_path=plot_path)
                print(f"📊 Training plot saved: {plot_path}")
            except Exception as e:
                print(f"⚠️  Could not save training plot: {e}")
        
        # Early stopping
        if early_stop_counter >= config.patience:
            print(f"\n🛑 EARLY STOPPING: No improvement for {config.patience} epochs")
            break
        
        # Memory cleanup
        memory_monitor.clear_cache()
        
        print(f"{'═' * 80}")

except KeyboardInterrupt:
    print(f"\n⚠️  Training interrupted by user")
except Exception as e:
    print(f"\n❌ Training error: {str(e)}")
    import traceback
    traceback.print_exc()

finally:
    total_training_time = time.time() - training_start_time
    
    print(f"\n🏁 TRAINING COMPLETED!")
    print("=" * 80)
    print(f"⏱️  Total Time: {format_time(total_training_time)}")
    print(f"🏆 Best Validation Dice: {best_val_dice:.4f}")
    print(f"📊 Epochs Completed: {len(training_history)}")
    print(f"💾 Model saved to: {Path(config.output_dir) / 'unet3d_model.pkl'}")
    print(f"🔥 RTX 4060 Performance: OPTIMIZED!")
    
    # Final plot
    try:
        final_plot_path = Path(config.output_dir) / "final_training_progress.png"
        visualizer.plot_training_progress(save_path=final_plot_path)
        print(f"📊 Final training plot: {final_plot_path}")
    except:
        pass
    
    # Save training history
    history_path = Path(config.output_dir) / "training_history.pkl"
    with open(history_path, 'wb') as f:
        pickle.dump(training_history, f)
    print(f"📈 Training history: {history_path}")
    
    print("=" * 80)
    
    # Memory cleanup
    memory_monitor.clear_cache()

# Print final model info
if 'model' in locals():
    print(f"\n🧠 FINAL MODEL INFORMATION:")
    print(f"├── Total Parameters: {sum(p.numel() for p in model.parameters()):,}")
    print(f"├── Model Size: {sum(p.numel() * 4 for p in model.parameters()) / 1024**2:.1f} MB")
    print(f"├── Best Validation Dice: {best_val_dice:.4f}")
    print(f"├── Architecture: {config.model_config.upper()} 3D U-Net")
    print(f"└── Ready for inference: ✅")

print(f"\n🎉 3D BRAIN TUMOR SEGMENTATION TRAINING COMPLETE!")
print(f"🔥 RTX 4060 OPTIMIZED | MIXED PRECISION | GPU ACCELERATED")
print(f"💾 Model ready for deployment!")
```
